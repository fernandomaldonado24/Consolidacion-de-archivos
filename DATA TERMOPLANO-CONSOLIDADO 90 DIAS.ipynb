```python
import pandas as pd
import matplotlib as plt
from pandas import ExcelWriter
import os
import openpyxl
import numpy as np
TC={ '1.2019':3.379,
     '2.2019':3.335,
     '3.2019':3.305,
     '4.2019':3.312,
     '5.2019':3.312,
     '6.2019':3.370,
     '7.2019':3.291,
     '8.2019':3.310,
     '9.2019':3.405,
     '10.2019':3.385,
     '11.2019':3.336,
     '12.2019':3.403,   
    '1.2022':3.998,
     '2.2022':3.846,
     '3.2022':3.759,
     '4.2022':3.701,
     '5.2022':3.838,
     '6.2022':3.707,
     '7.2022':3.830,
     '8.2022':3.925,
     '9.2022':3.847,
     '10.2022':3.984,
     '11.2022':3.993,
     '12.2022':3.854,
     '1.2023':3.820,
     '2.2023':3.851}
input_loc='//192.168.10.250/finanzas_produccion/FINANZAS_PRODUCCION/Fernando/BASE DE DATOS/DATA DIARIA/DATA LIMPIA/'
output_loc='//192.168.10.250/finanzas_produccion/FINANZAS_PRODUCCION/Fernando/BASE DE DATOS/DATA TERMOPLANO/'
df1=pd.DataFrame()
finalDF=pd.DataFrame()
FECHA_BASE=str(input('INGRESA LA FECHA BASE'))
#       FECHA FINAL
AÑO=int(FECHA_BASE[-4:])
MES=int(FECHA_BASE[3:5])
DIA=int(FECHA_BASE[:2])
#       FECHA INICIAL
#CALCULAR CANTIDAD DE DIAS DEL MES
def es_bisiesto(año: int):
    return (año-2020)%4 == 0

def cant_dias(año:int,mes:int):
    if mes in [4,6,9,11]:
        return 30
    if mes==2:
        if es_bisiesto(año):
            return 29
        else:
            return 28
    else:
        return 31

def siguiente_mes(año:int,mes:int):
    if mes-3<=0:
        if mes==3:
            x= int(DIA+1+cant_dias(año,2)+cant_dias(año,1)+cant_dias(año-1,12)-90)
            if x>cant_dias(AÑO,12):
                return 1
            else:
                return 0
        elif mes==2:
            x= int(DIA+1+cant_dias(año,1)+cant_dias(año-1,12)+cant_dias(año-1,11)-90)
            if x>cant_dias(AÑO,11):
                return 1
            else:
                return 0
        elif mes==1:
            x= int(DIA+1+cant_dias(año-1,12)+cant_dias(año-1,11)+cant_dias(año-1,10)-90)
            if x>cant_dias(AÑO,10):
                return 1
            else:
                return 0
    else:
        x= int(DIA+1+cant_dias(año,MES-1)+cant_dias(año,MES-2)+cant_dias(año-1,MES-3)-90)
        if x>cant_dias(AÑO,MES-3):
            return 1
        else:
            return 0 

def MES_INICIAL(mes:int):
    if mes-3<=0:
        if mes==3:
            if siguiente_mes(AÑO,MES)==1:
                return 1
            else:
                return 12
        elif mes==2:
            if siguiente_mes(AÑO,MES)==1:
                return 12
            else:
                return 11
        elif mes==1:
            if siguiente_mes(AÑO,MES)==1:
                return 11
            else:
                return 10
    else:
        if siguiente_mes(AÑO,MES)==1:
            return mes-2
        else:
            return mes-3

def DIA_INICIAL(año:int,mes:int):
    if mes-3<=0:
        if mes==3:
            x= int(DIA+1+cant_dias(año,2)+cant_dias(año,1)+cant_dias(año-1,12)-90)
            if x>cant_dias(AÑO,12):
                return x-cant_dias(AÑO,12)
            else:
                return x
        elif mes==2:
            x= int(DIA+1+cant_dias(año,1)+cant_dias(año-1,12)+cant_dias(año-1,11)-90)
            if x>cant_dias(AÑO,11):
                return x-cant_dias(AÑO,11)
            else:
                return x
        elif mes==1:
            x= int(DIA+1+cant_dias(año-1,12)+cant_dias(año-1,11)+cant_dias(año-1,10)-90)
            if x>cant_dias(AÑO,10):
                return x-cant_dias(AÑO,10)
            else:
                return x
    else:
        x= int(DIA+1+cant_dias(año,MES-1)+cant_dias(año,MES-2)+cant_dias(año-1,MES-3)-90)
        if x>cant_dias(AÑO,MES-3):
            return x-cant_dias(AÑO,MES-3)
        else:
            return x


def AÑO_INICIAL(año:int,mes:int):
    if mes-3<=0:
        return año-1
    else:
        return año

#       CALCULO DE AÑO, MES Y DIA
DIA_INICIAL=DIA_INICIAL(AÑO,MES)
MES_INICIAL=MES_INICIAL(MES)
AÑO_INICIAL=AÑO_INICIAL(AÑO,MES)
FECHA_INICIAL=str(DIA_INICIAL)+'.'+str(MES_INICIAL)+'.'+str(AÑO_INICIAL)
#print(FECHA_INICIAL+'-'+FECHA_BASE)
print(str(AÑO_INICIAL)+str(MES_INICIAL)+str(DIA_INICIAL))
```
